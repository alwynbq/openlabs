```python
import matplotlib.pyplot as plt
```

```python
x = [0,2,4,6,8,10]
y = [20,15,35,40,41,47]

z = [50,70,80,40,90,100]
```

```python
plt.plot(x,y)

plt.title('x vs y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(x,y, color = 'g', marker='o', linestyle='-.') 
plt.plot(x,z, color = 'r', marker='x', linestyle='-.')
plt.bar(x,y)
plt.title('x vs y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend (['y','z'])
plt.grid(True)
plt.show()
```
